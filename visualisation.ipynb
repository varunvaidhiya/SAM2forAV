```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame with the benchmarking results
data = {
    'Model': ['SegFormer', 'Original SAM2', 'Fine-tuned SAM2', 'Optimized SAM2'],
    'Mean IOU (%)': [49.00, 15.08, 26.70, 2.22],
    'Mean Precision (%)': [None, 100.00, 100.00, 49.92],
    'Mean Recall (%)': [None, 15.08, 26.70, 28.40],
    'Mean F1 Score (%)': [None, 23.29, 30.48, 4.20],
    'Mean Pixel Accuracy (%)': [69.00, 15.11, 26.71, None],
    'Mean Latency (ms)': [129.0, 4.94, 4.77, 2.02],
    'FPS': [7.75, None, None, None],
}

df = pd.DataFrame(data)
df.set_index('Model', inplace=True)

# 1. Mean IOU Comparison
plt.figure(figsize=(8, 6))
df['Mean IOU (%)'].plot(kind='bar', color='skyblue')
plt.ylabel('Mean IOU (%)')
plt.title('Mean IOU Comparison Across Models')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Precision, Recall, and F1 Score Comparison
metrics = ['Mean Precision (%)', 'Mean Recall (%)', 'Mean F1 Score (%)']
df_metrics = df[metrics].dropna()

df_metrics.plot(kind='bar', figsize=(10, 6))
plt.ylabel('Percentage (%)')
plt.title('Precision, Recall, and F1 Score Comparison Across Models')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

# 3. Mean Pixel Accuracy Comparison
plt.figure(figsize=(8, 6))
df['Mean Pixel Accuracy (%)'].dropna().plot(kind='bar', color='green')
plt.ylabel('Mean Pixel Accuracy (%)')
plt.title('Mean Pixel Accuracy Comparison Across Models')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Mean Inference Latency Across Models
plt.figure(figsize=(8, 6))
df['Mean Latency (ms)'].plot(kind='bar', color='orange')
plt.ylabel('Mean Latency (ms)')
plt.title('Mean Inference Latency Across Models')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. FPS for SegFormer
if not df['FPS'].isnull().all():
    plt.figure(figsize=(6, 4))
    df['FPS'].dropna().plot(kind='bar', color='purple')
    plt.ylabel('Frames Per Second')
    plt.title('FPS for SegFormer')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
